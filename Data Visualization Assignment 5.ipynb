#### You have been provided with data from the Titanic Passenger Roaster, depicting if an individual survived the sinking of the ship or not. ‘Survived’ is denoted by (1) and ‘Did Not Survive’ is denoted by (0). Only using Plotly, graph the following information from the data. Feel free to use any of the Plotly submodules discussed (graph_objects, plotly express or iplot). Remember to label your axis and name your plots appropriately. For each given question, only one figure is expected.

```python
import pandas as pd
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
```

Load the Dataset

```python
titanic = pd.read_csv('Titanic Data.csv')
```

```python
titanic.head()
```

***

Qn 1: Show the age distribution in the data using a histogram

```python
df_age = titanic.Age
df_age.head()
```

```python
trace = go.Histogram(
    x=df_age,
    opacity=0.75,
    name = "Age",
    marker=dict(color='violet'))#rgba(171, 50, 96, 0.6)

layout = go.Layout(barmode='overlay',
                   title=' Age distribution Histogram',
                   xaxis=dict(title='Age'),
                   yaxis=dict( title='Count'),
)
fig = go.Figure(data=trace, layout=layout)

fig.show()
```

Qn 4: Pivot the data setting the column Pclass as the columns and Fare as the values. From the resulting structure, use a boxplot to show the distribution of the values in its 3 columns

```python
df_pivot = titanic.pivot(columns='Pclass', values='Fare') #index='', 
```

```python
df_pivot.head()
```

```python
trace0 = go.Box(
    y=df_pivot[1],
    name = 'Pclass 1',
    marker = dict(
        color = 'red',#rgb(12, 12, 140)
    )
)
trace1 = go.Box(
    y=df_pivot[2],
    name = 'Pclass 2',
    marker = dict(
        color = 'Blue',#rgb(12, 128, 128)
    )
)
trace2 = go.Box(
    y=df_pivot[3],
    name = 'Pclass 3',
    marker = dict(
        color = 'magenta',#rgb(12, 128, 128)
    )
)
data = [trace0, trace1, trace2]


fig = go.Figure(data=data)


#fig = go.Figure(dict(data = data, layout = layout))

fig.show()
```
